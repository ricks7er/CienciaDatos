Importar librerias

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

Cargar el dataset de prueba

```python
infancia = pd.read_csv('C:\Users\Estudiantes\Downloads\infanciaAdoles.csv', sep=',')
infancia
```

isnull() -> Indica si una celda esta vacia en el dataset

```python
infancia_nan = infancia.isnull()
infancia_nan
```

dropna(how='any'), Muestra las filas que no tienen ningun campo vacio, en caso de querer todos los elementos de la fila vacios, el parametro how es igual a 'all'

```python
infancia_sin_nan = infancia.dropna(how='any')
infancia_sin_nan
```

fillna(value = {
    '2010':0,
    '2011':0,
    '2012':0,
    '2013':0,
    '2014':0, '2015':0,
    'ESTADO': 'No calculable',
    'OBSERVACIONES': 'No aplica',
    'FUENTE': 'Sin fuente'
}) -> Cambia los campos vacios por columna por el especificado

```python
infancia_nuevo = infancia.fillna(value = {
    '2010':0,
    '2011':0,
    '2012':0,
    '2013':0,
    '2014':0, '2015':0,
    'ESTADO': 'No calculable',
    'OBSERVACIONES': 'No aplica',
    'FUENTE': 'Sin fuente'
})
infancia_nuevo
```

fillna(method='bfill', limit=1) -> Copia el valor proximo a un vacio a la celda vacia

```python
infancia_nuevo2 = infancia.fillna(method='bfill', limit=1)
infancia_nuevo2
```

infancia['2010'] >= 5 Retorna una serie de valores booleanos, False si no cumple la condición y True si la cumple 

```python
infancia['2010'] >= 5
```

infancia[infancia['2010'] >= 5] Retorna el dataframe de las filas que cumplen la condicion dentro de las llaves

```python
infancia[infancia['2010'] >= 5]
```

infancia['Nombre del Indicador'][infancia['2010'] >= 5] Devuelve el serie de un dataframe que cumple la condicion

```python
infancia['Nombre del Indicador'][infancia['2010'] >= 5]
```

infancia[(infancia['2010'] >= 5) & (infancia['2011'] >= 200)] Retorna el dataframe de las filas que cumplen las condiciones dentro de las llaves

```python
infancia[(infancia['2010'] >= 5) & (infancia['2011'] >= 200)]
```

Where mantiene los valores en una serie que cumplan la condicion, los que no por defecto les asignan NaN. El segundo argumento es el valor para los elementos de la serie que no cumplen la condicion

```python
infancia['ESTADO'].where(infancia['ESTADO'] == 'MEJORO')
```

Mask mantiene los valores en una serie que no cumplan la condicion, los que si por defecto les asignan NaN.

```python
infancia['ESTADO'].mask(infancia['ESTADO'] == 'MEJORO')
```

isin Retorna un dataframe de valores booleano, True para los campos que concuerden con algun elemento del argumento o False en caso contrario

```python
infancia.isin(['MEJORO'])
```

query -> Consulta mediante un cadena de texto con operaciones booleanas dadas sobre las columnas

```python
infancia = infancia.rename(columns={'2010': "a", '2011': "b", '2012': "c", '2013': "d", '2014': "e"})
```
